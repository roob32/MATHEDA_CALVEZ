Import <b>pandas</b>, <b>numpy</b> and <b>matplotlib</b> library

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as mpl
```

```python
df = pd.read_csv(r"..\.raw\chip-dataset-cleaned.csv", delimiter=",")
df.info()
```

```python
df = df.drop(df.columns[[10, 11, 12]], axis=1 )
df.head()
```

## N Chips Manufactured

```python
df_w = df.copy()
```

```python
chipCount = df_w['Vendor'].value_counts()
chipCount
```

```python
type(chipCount)
```

```python
chart = chipCount.plot(kind='bar', color=['#efbf04', '#efbf04', '#efbf04', '#000000', '#000000'], figsize=(12, 6))
chart.bar_label(chart.containers[0], label_type='edge')
mpl.title('Chips Manufactured (CPU + GPU)', 
          fontfamily='Arial', 
          fontsize=11, 
          fontweight='bold')
mpl.xlabel('')
chart.set_xticklabels(chipCount.index, rotation=0)
mpl.ylabel('N Chips')
mpl.show()
```

## N CPU Manufactured

```python
df_cpu = df_w[df_w['Type']=='CPU']
df_cpu
```

```python
cpuCount = df_cpu['Vendor'].value_counts()
cpuCount
```

```python
chart = cpuCount.plot(kind='bar', color='#efbf04', figsize=(8,4))
mpl.title('CPU Manufactured', 
          fontfamily='Arial', 
          fontsize=11, 
          fontweight='bold')
mpl.ylabel('N Count')
mpl.xlabel('')
mpl.show()
```

## Lowest TDP (Thermal Design Power)

```python
df_tdp = df_w[['Product', 'TDP']]
df_tdp
```

```python
type(df_tdp)
```

```python
df_tdp = df_tdp.sort_values(by='TDP')
df_tdp
```

```python
df_tdp = df_tdp.head(8)
df_tdp
```

```python
mpl.figure(figsize=(12,6))
mpl.bar(df_tdp['Product'],df_tdp['TDP'], color='#efbf04')
mpl.title('Products with the Lowest TDP', 
          fontfamily='Arial', 
          fontsize=11, 
          fontweight='bold')
mpl.ylabel('Watts')
mpl.xticks(rotation=45, ha='right')
mpl.show()
```

## GPU Process Size Trend

```python
df_ps = df_w[['Type','Date','Process_Size']].where(df_w['Type']=='GPU').dropna()
df_ps
```

```python
df_ps['Date'] = pd.to_datetime(df_ps['Date'])
df_ps['Year'] = df_ps['Date'].dt.year
df_ps 
```

```python
df_ps = df_ps.groupby('Year')['Process_Size'].mean().reset_index()
df_ps
```

```python
mpl.figure(figsize=(10,5))
mpl.plot(df_ps['Year'],df_ps['Process_Size'], color='#efbf04')
mpl.title('GPU Process Size Trend')
mpl.ylabel('nm')
```

## Conclusion

<p>Over the years, the GPU Processor Size is slowly reaching a stable point somewhere close to 0, indicating the end of Moore's Law which states that as transistors get smaller, it becomes increasingly more difficult to print them. The number of Chips produced over the years also supports this claim since the number of chips that were produced also decreased, indicating a larger timeframe required to manufacture chips of smaller size.</p>
