## Setup

Import <b>Pandas</b>, <b>numpy</b>, and <b>matplotlib</b> library

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as mpl
```

# Trend Related Analysis

## Enrollment Trend Over Time

```python
df_enrollment = df.copy()
df_enrollment = df_enrollment[['YEAR', 'ENROLL']].dropna()
df_enrollment = df_enrollment.groupby('YEAR')['ENROLL'].mean().reset_index()
df_enrollment
```

```python
mpl.figure(figsize=(10,6))
mpl.plot(df_enrollment['YEAR'], df_enrollment['ENROLL'], color='#efbf04')
mpl.title('Enrollment Trend Over Time')
mpl.ylabel('Total Student Enrollment per year')
mpl.show()
```

## Per-Student Expenditure Analysis

```python
df_studexp = df.copy()
df_studexp = df_studexp[['STATE', 'YEAR', 'ENROLL', 'TOTAL_EXPENDITURE']].dropna()
df_studexp['EXPENDITURE_PER_STUDENT'] = df_studexp['TOTAL_EXPENDITURE'] / df_studexp['ENROLL']
df_studexp.head(10)
```

### Per-Student Expenditure Over Time

```python
df_studexp1 = df_studexp.copy()
df_studexp1 = df_studexp.groupby('YEAR')['EXPENDITURE_PER_STUDENT'].mean().reset_index() 
```

```python
mpl.figure(figsize=(10,6))
mpl.plot(df_studexp1['YEAR'], df_studexp1['EXPENDITURE_PER_STUDENT'], color='#efbf04')
mpl.title('Per-Student Expenditure Over Time')
mpl.ylabel('Average Expenditure Per-Student Per Year')
mpl.show()
```

### Highest & Lowest Per-Student Expenditure

```python
df_studexp_high = df_studexp.copy()
df_studexp_high = df_studexp_high[df_studexp_high['YEAR'] == df_studexp_high['YEAR'].max()]
df_studexp_high = df_studexp.groupby('STATE')['EXPENDITURE_PER_STUDENT'].mean().reset_index()
df_studexp_high = df_studexp_high.sort_values(by='STATE', ascending=True)
df_studexp_high.head(10)
```

```python
mpl.figure(figsize=(18,8))
bars1 = mpl.bar(df_studexp_high['STATE'], df_studexp_high['EXPENDITURE_PER_STUDENT'], color='#efbf04', linewidth=3)
mpl.title('Per-Student Expenditure by State (Latest Year)', fontsize=20)
mpl.xlabel('Expenditure Per Student (USD)', fontsize=20)
mpl.xticks(df_studexp_high['STATE'], rotation=90 , fontsize=14)
mpl.ylabel('State' , fontsize=20)

for bar in bars1:
    mpl.text(bar.get_x() + bar.get_width()/2, bar.get_height(), 
             f'{bar.get_height():.1f}', ha='center', va='bottom', fontsize=10)
mpl.grid(axis='y', color='#4a4a4a', linestyle='--', linewidth='0.5', alpha=0.5)
mpl.show()
```
